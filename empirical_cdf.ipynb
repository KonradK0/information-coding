```
import sympy
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interpolate


x = sympy.Symbol('x')
rho = 1-4*x+6*x**2
integral = sympy.integrate(rho, x)
integral
```

```
sample_size = 1000

xs = np.linspace(0, 1, num=sample_size)
ys = []
integrals = []
for x in xs:
  ys.append(rho.evalf(subs={'x':x}))
  integrals.append(integral.evalf(subs={'x' : x}))
  
n = 200
rev = interpolate.interp1d(integrals, xs)
uni = np.random.random(size=n)
pts = np.sort(rev(uni))
cdfe = []

for i, _ in enumerate(pts):
  cdfe.append((i-.5)/n)
cdfe = np.array(cdfe)
```

```
plt.plot(xs, ys)
plt.plot(xs, integrals)
plt.scatter(uni, np.zeros_like(uni), marker="|", color='y')
plt.scatter(pts, cdfe)

fig = plt.gcf()
fig.set_size_inches(10, 8)
plt.show()
```
